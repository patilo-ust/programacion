# Cajero Express
Instrucciones: Simula un menú de cajero automático.
Define una variable saldo = 1000.

Muestra el siguiente menú:

1. Ver Saldo

2. Retirar Dinero

3. Salir

Reglas:

1: Mostrar el valor de la variable saldo.

2: Imprimir "Procesando retiro de efectivo...".

3: Imprimir "Gracias por usar nuestro cajero. ¡Adiós!".

(Default): Imprimir "Opción no válida, intente de nuevo".

```python
#aqui bien nos indican que es un menu de opciones, por ende si o si debe ser match
#podemos pedirle al usuario(en este caso) que ingrese el salgo o dejarlo por default
#a codear!
saldo = float(input("ingrese su saldo: $...."))
print(" bienvenido a su cajero virtual")
print("opcion 1: ver saldo: ")
print("opcion 2: retirar dinero: ")
print("opcion 3: salir: ")
opcion = int(input(" ingrese una opción "))
```

```python
#ya tengo mis variables y datos, ahora hagamos el menu
match opcion:
  case 1:
    print(f"su saldo es:{saldo}")
  case 2:
    saldo2 = float(input(f" su retiro maximo es: ${saldo}, cuanto desea retirar?: "))
    retiro = saldo-saldo2
    print(f" se ha retirado: ${saldo2}, su nuevo saldo es: $ {retiro}")
  case 3:
    print("hasta pronto")
  case _:
    print("opcion no valida... ")

```

# Ejercicio 5: Sistema de Cobro de Estacionamiento Profesional Diseña un programa con el siguiente flujo:
# Pregunta al usuario cuántas horas permaneció el auto.
#Si las horas son menores o iguales a 0, muestra un error.
#Selección de Vehículo: Si las horas son válidas, pregunta el tipo de vehículo:
# “Auto”: El costo es $5 por hora.  

# “Camión”: El costo es $10 por hora.

# "MOTO": costo es $2 por hora.

#Validación Interna Dentro de cada opción (por ejemplo, dentro de "Auto"), verifica:
#•	Si las horas son más de 8, aplica un descuento del 10% al total.
#•	Si no, muestra el total normal.



```python
#lo que haremos es pedir datos y crear variables, tambien podemos crear los calculos enseguida
horas = float(input("ingrese la cantidad de horas que lleva: "))
vehiculo = input("ingrese el tipo de vehiculo: (auto, moto o camion) ").lower()
#total sin descuento por cada vehiculo
sindcto_auto = horas * 5
sindcto_camion = horas * 10
sindcto_moto = horas * 2
#total con descuento
cdcto_auto= sindcto_auto*0.90
cdcto_camion= sindcto_camion*0.90
cdcto_moto= sindcto_moto * 0.90


```

```python
#ahora sabemos que debe ser 1 tipo de vehiculo, podemos hacer un match segun la opcion y luego revisar las horas

match vehiculo:
  case "auto":
    print(f"su vehiculo es un: {vehiculo} y tiene: {horas} horas")
    if horas >= 1 and horas <=7:
      print(f"ud paga: ${sindcto_auto}")
    elif horas >=8:
      print(f"ud tiene un descuento de 10%, debe pagar: ${cdcto_auto}")
    else:
      print("horas no validas")
  case "moto":
    print(f"su vehiculo es un: {vehiculo} y tiene: {horas} ")
    if horas >= 1 and horas <=7:
      print(f"ud paga: ${sindcto_moto}")
    elif horas >=8:
      print(f"ud tiene un descuento de 10%, debe pagar: ${cdcto_moto}")
    else:
      print("horas no validas")
  case "camion":
    print(f"su vehiculo es un: {vehiculo} y tiene: {horas} ")
    if horas >= 1 and horas <=7:
      print(f"ud paga: ${sindcto_camion}")
    elif horas >=8:
      print(f"ud tiene un descuento de 10%, debe pagar: ${cdcto_camion}")
    else:
      print("horas no validas")
  case _:
    print("vehiculo no encontrado")
```

tambien es posible acotar el codigo de la siguiente mantera

```python
match vehiculo:
    case "auto":
        tarifa = 5
    case "moto":
        tarifa = 2
    case "camion":
        tarifa = 10
    case _:
        print("Vehículo no válido")
        exit()

total = horas * tarifa

if horas > 8:
    total *= 0.9
    print(f"Tiene descuento. Total a pagar: ${total}")
else:
    print(f"Total a pagar: ${total}")
```
